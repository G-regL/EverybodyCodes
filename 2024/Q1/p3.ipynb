```python
# data = "xBxAAABCDxCC"
data = open("input_p3.txt").readlines()[0]
creatures = {"A": 0, "B": 1, "C": 3, "D": 5}

total_potions = 0
battle_size = 3

for idx in range(0, len(data), battle_size):
    potions = 0
    potions_buff = 0
    battle_creatures = data[idx : idx + battle_size]
    print(f"Battle: {battle_creatures}")
    battle_potions = [
        battle_creatures.count(c) * p
        for c, p in creatures.items()
        if c in battle_creatures
    ]
    potions += sum(battle_potions)
    print(f"  Potions: {potions}; {battle_potions}")
    spaces = battle_creatures.count("x")

    # need to account for P2 stuff.. good grief..
    if spaces == 1:
        potions_buff = 2
        print(f"  Buffing {potions_buff} ({spaces})!!")
    if spaces == 0:
        potions_buff = 6
        print(f"  Buffing {potions_buff} ({spaces})!!")

    total_potions += potions + potions_buff
    print(f"  Added {potions + potions_buff} potion(s) to total ({total_potions})")


total_potions
```
