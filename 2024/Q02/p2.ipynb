```python
from IPython.core.interactiveshell import InteractiveShell

InteractiveShell.ast_node_interactivity = "all"

data_raw = """WORDS:THE,OWE,MES,ROD,HER,QAQ

AWAKEN THE POWE ADORNED WITH THE FLAMES BRIGHT IRE
THE FLAME SHIELDED THE HEART OF THE KINGS
POWE PO WER P OWE R
THERE IS THE END
QAQAQ"""
data = data_raw.split("\n")
data = open("input_p2.txt").read().split("\n")

words = data[0].split(":")[1].split(",")
inscriptions = data[2:]

words.extend([w[::-1] for w in words if w[::-1] not in words])

symbols = 0
found = set()

for ins in inscriptions:
    ins_symbols = [0 for _ in range(len(ins))]
    for word in words:
        for idx in range(0, len(ins) - len(word) + 1):
            if ins[idx:].startswith(word):
                ins_symbols = ins_symbols[:idx] + [1 for _ in range(len(word))] + ins_symbols[idx + len(word) :]

    symbols += sum(ins_symbols)

symbols
```
