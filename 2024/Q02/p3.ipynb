```python
from IPython.core.interactiveshell import InteractiveShell

InteractiveShell.ast_node_interactivity = "all"

data_raw = """WORDS:THE,OWE,MES,ROD,RODEO

HELWORLT
ENIGWDXL
TRODEOAL"""
data = data_raw.split("\n")
data = open("input_p3.txt").read().split("\n")

words = data[0].split(":")[1].split(",")
words.extend([w[::-1] for w in words if w[::-1] not in words])
shield = {(r, c): col for r, row in enumerate(data[2:]) for c, col in enumerate(row)}

width = len(data[2])

scales = set()

# loop through the words
for word in words:
    # Loop through each character in the shield
    for location, char in shield.items():
        # Check if we found a possible word start
        if word[0] == char:
            # forward look
            # track how many chars we've found for this word while looking right
            chars_found = {(location[0], location[1])}
            # Looking forward as far as the word is long
            for distance in range(1, len(word)):
                look = (location[0], location[1] + distance)
                # If we're off the right of the shield, wrap around to the left
                if look not in shield:
                    look = (location[0], location[1] - width + distance)
                # If where we're looking matches the word character, add it to the set
                if shield[look] == word[distance]:
                    chars_found.add(look)
            # If the amount of chars found matches the length of the word, add all those locations to the scales set
            if len(chars_found) == len(word):
                scales.update(chars_found)

            # down look
            # track how many chars we've found for this word while looking down
            chars_found = {(location[0], location[1])}
            # Looking down as far as the word is long
            for distance in range(1, len(word)):
                look = (location[0] + distance, location[1])
                # If we're not off the bottom of the shield, and the char matches, add it to the set
                if look in shield and shield[look] == word[distance]:
                    chars_found.add(look)
            # If the amount of chars found matches the length of the word, add all those locations to the scales set
            if len(chars_found) == len(word):
                scales.update(chars_found)


len(scales)
```
