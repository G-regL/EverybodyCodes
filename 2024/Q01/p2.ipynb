```python
# data = "AxBCDDCAxD"
data = open("input_p2.txt").readlines()[0]
creatures = {"A": 0, "B": 1, "C": 3, "D": 5}

total_potions = 0

for idx in range(0, len(data), 2):
    potions = 0
    battle_creatures = data[idx : idx + 2]
    # print(f"Battle: {battle_creatures}")
    battle_potions = [
        battle_creatures.count(c) * p
        for c, p in creatures.items()
        if c in battle_creatures
    ]
    potions += sum(battle_potions)
    # print(f"  Potions: {potions}; {battle_potions}")
    if not battle_creatures.count("x"):
        # print("  Buffing!!")
        potions += 2

    # print(f"  Adding {potions} potions to total ({total_potions})")
    total_potions += potions

total_potions
```
