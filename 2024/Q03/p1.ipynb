```python
from IPython.core.interactiveshell import InteractiveShell

InteractiveShell.ast_node_interactivity = "all"

data = """..........
..###.##..
...####...
..######..
..######..
...####...
.........."""
data = data.split("\n")
data = open("input_p1.txt").read().split("\n")

data = [l.replace("#", "1") for l in data]
grid = {(row, col): char for row, line in enumerate(data) for col, char in enumerate(line)}
# grid

directions = [(0, -1), (-1, 0), (0, 1), (1, 0)]

current_depth = 1
while True:
    changes = []
    for (row, col), char in grid.items():
        # get surrounding cells, or "." if it's our of range
        surrounding2 = [grid.get((row + d[0], col + d[1]), ".") for d in directions]

        # check that all the surrounding cells are greater than the curren depth
        # Need to also check that they're digits, which removes the "."
        if all(s.isdigit() and int(s) >= current_depth for s in surrounding2):
            grid[(row, col)] = str(int(grid[(row, col)]) + 1)
            changes.append((row, col))

    if not changes:
        break
    current_depth += 1


sum([int(char) for char in grid.values() if char != "."])

for row in range(len(data)):
    for col in range(len(data[0])):
        cell = grid[(row, col)]
        print(cell, end="")
    print()
print()
```

```python
from IPython.core.interactiveshell import InteractiveShell

InteractiveShell.ast_node_interactivity = "all"

data = """..........
..###.##..
...####...
..######..
..######..
...####...
.........."""
data = data.split("\n")
data = open("input_p1.txt").read().split("\n")

data = [l.replace("#", "1") for l in data]
grid = {(row, col): char for row, line in enumerate(data) for col, char in enumerate(line)}
# grid

directions = [(0, -1), (-1, 0), (0, 1), (1, 0)]
# blocks = 0
for (row, col), char in grid.items():
    if char == ".":
        continue
    print(f"{row, col} {char}")
    distance = 1
    # block_depth = 1
    found_point = False
    while not found_point:
        print(f" Looking at {distance=}")
        # block_depth += 1
        for dir in directions:
            look = (row + distance * dir[0], col + distance * dir[1])

            if look not in grid:
                continue

            if grid[look] == ".":
                print("  Found '.'")
                found_point = True
                block_depth = max(abs(look[0] - row), abs(look[1] - col))
                break

        distance += 1

    # blocks += block_depth
    grid[(row, col)] = block_depth
    print(f"  max_depth for point is {block_depth}")
    # print(f"  {blocks=}")


# blocks

# #grid
sum([char for char in grid.values() if char != "."])
# sum([char for char in grid.values() if char != "."])
# sum([1 for char in grid.values() if isinstance(char,int) and char >= 2])
# sum([1 for char in grid.values() if isinstance(char,int) and char >= 3])
# sum([1 for char in grid.values() if isinstance(char,int) and char >= 4])
# sum([1 for char in grid.values() if isinstance(char,int) and char >= 5])

for row in range(len(data)):
    for col in range(len(data[0])):
        cell = grid[(row, col)]
        print(cell, end="")
    print()
print()
```
